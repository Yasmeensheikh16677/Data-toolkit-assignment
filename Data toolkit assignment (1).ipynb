```python
#3.Explain the following terms
1) The difference in np.array, np.asarray and np.asanyarray
2) The difference between Deep copy and shallow copy
```

```python
# 1) np.array: it creates a copy of the object array or the original array and does not reflect any changes made to the original array.syntax: np.asanyarray(arr)
#np.asarray:it would reflect all the changes made to the original array. syntax: np.asarray(arr)
#np.asanyarray: The numpy.asanyarray() function is used to convert the input to an ndarray, but pass ndarray subclasses through.syntax: np.asanyarray(arr)
```

```python
#2) Deep copy
l=[1,2,3,4]
arr=np.array(l)
print ("Array:", arr)
arr[1]=100
print("Array after changing values at 1th place:", arr)
new_array=arr
print("New array as:", new_array)
arr[1]=2000
print("New array after change in original array:", new_array)
```

```python
#Shallow copy
l=[1,2,3,4]
arr=np.array(l)
print ("Array:", arr)
a=arr.copy()
print ("Copied Array:", a)
arr[2]=300
print("Array after changing 2nd value:", arr)
print("Copied array:", a)
```

```python
#17.Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
x=np.random.randn(500)
y=np.random.randn(500)
quadrant = np.zeros_like(x, dtype=int)
quadrant[(x > 0) & (y > 0)] = 1  # Quadrant 1
quadrant[(x < 0) & (y > 0)] = 2  # Quadrant 2
quadrant[(x < 0) & (y < 0)] = 3  # Quadrant 3
quadrant[(x > 0) & (y < 0)] = 4  # Quadrant 4
plt.figure(figsize=(8, 6))
sns.scatterplot(x,y,style='smoker', palette='Dark2', legend='full')
plt.legend(title='Quadrant')
plt.title('Quadrant-wise Scatter Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
#19.Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their values, add hover tooltips to display exact values, label the axes, and set the title as 'Random CategoricalBar Chart
```

```python
import random
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
import numpy as np
from bokeh.plotting import figure, output_file, show
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.transform import factor_cmap

```

```python
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
data = {'categories': categories, 'values': [random.randint(1, 100) for _ in range(len(categories))]}
source = ColumnDataSource(data=data)
p = figure(x_range=categories, title="Random Categorical Bar Chart",toolbar_location=None, tools="")
p.vbar(x='categories', top='values', width=0.9, source=source,
line_color='white', fill_color=factor_cmap('categories', palette=['red', 'green', 'blue', 'orange', 'purple'], factors=categories))
hover = HoverTool()
hover.tooltips = [("Value", "@values")]
p.add_tools(hover)
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.yaxis.axis_label = 'Values'
p.xaxis.axis_label = 'Categories'
show(p)
```
